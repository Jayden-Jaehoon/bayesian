### 1. **완전 조건부 분포와 깁스 샘플링**

#### 2.2. 완전 조건부 분포와 깁스 샘플링

- $\beta$의 사전 분포는 $\beta | \sigma^2 \sim Normal(\beta_0, \sigma^2 B_0)$로 설정한다. 이는 $\sigma^2$에 의존하는데, 이는 연구자의 사전 정보에 따른 것이다. 따라서 사전 분산이 작을수록 사후 분포에 강하게 반영된다.
- 사전 정보가 많을수록 추정값의 편차가 줄어들며, 이는 실무적으로 중요한 의미를 가진다.
- 연구자가 추정값을 더 유동적으로 설정하려면 $\beta$에 독립된 사전 분포로 $\beta \sim Normal(\beta_0, B_0)$를 사용할 수 있다.
  
#### 선형 회귀모형의 설정
- $\sigma^2 \sim InverseGamma\left(\frac{\alpha_0}{2}, \frac{\delta_0}{2}\right)$
- $\beta \sim Normal(\beta_0, B_0)$
- $Y|\beta, \sigma^2 \sim Normal(X\beta, \sigma^2 I_T)$

---

### 2. **사례 A: $\sigma^2$가 알려진 경우**

- 이 경우 $\beta$의 사후 분포는 분산이 $B_1 = \left(\sigma^{-2}X'X + B_0^{-1}\right)^{-1}$이고, 평균이 $B_1(\sigma^{-2}X'Y + B_0^{-1}\beta_0)$인 정규 분포이다.
- $B_1 = \left(\sigma^{-2}X'X + B_0^{-1}\right)^{-1}$
- $A = \sigma^{-2}X'Y + B_0^{-1} \beta_0$
- $\beta | Y, \sigma^2 \sim Normal(B_1 A, B_1)$

### 3. **사례 B: $\beta$가 알려진 경우**

- $\sigma^2$의 사후 분포는 $\beta$의 사전 분포와 $\sigma^2$에 의존하며, 다음과 같이 도출된다.
- $\sigma^2 | Y, \beta \sim InverseGamma\left(\frac{a_1}{2}, \frac{\delta_1}{2}\right)$
- 여기서 $a_1 = a_0 + T$, $\delta_1 = \delta_0 + (Y - X\beta)'(Y - X\beta)$

### 4. **사례 C: $\beta$와 $\sigma^2$ 모두 알려지지 않은 경우**

- $\beta$와 $\sigma^2$에 결합된 사후 분포는 우도 함수와 사전 밀도의 곱에 비례한다:
  $
  \pi(\beta, \sigma^2 | Y) \propto f(Y|\beta, \sigma^2) \pi(\beta|\sigma^2) \pi(\sigma^2)
  $
  이를 풀면 $\beta, \sigma^2 | Y$의 결합 사후 분포는 다음과 같다:
  $
  \pi(\beta, \sigma^2 | Y) = Normal(Y | X\beta, \sigma^2) \times Normal(\beta | \beta_0, \sigma^2 B_0) \times InverseGamma\left(\frac{\alpha_0}{2}, \frac{\delta_0}{2}\right)
  $
- 결합 사후 분포는 표준화된 형태로 정리된다.

---

### 5. **완전 조건부 분포**

- 경우 C와 같이 $\beta, \sigma^2$의 사후 분포를 추출할 수 없는 경우, 깁스 샘플링을 통해 접근할 수 있다. 
- 이 방법은 단일 조건부 분포로부터 각각의 매개변수를 추정하고, 이 과정을 반복해 사후 분포를 추정하는 방식이다.

---

### 6. **깁스 샘플링의 구조**

- 깁스 샘플링은 조건부 분포로부터 샘플링하여 이를 반복하는 방식이다.
- 반복적으로 $\beta$와 $\sigma^2$를 번갈아가며 샘플링한다:
  - $\beta^{(n+1)} \sim p(\beta | \sigma^{2(n)}, Y)$
  - $\sigma^{2(n+1)} \sim p(\sigma^2 | \beta^{(n+1)}, Y)$
- 이 과정에서 첫 번째 $n_0$개의 샘플들은 버려지는 "Burn-in" 과정이 필요하다. 

---

### 7. **사후 분포 수렴**

- 사후 분포의 수렴을 확인하기 위해 초기 $n_0$개의 샘플은 버려지며, 나머지 샘플을 분석에 사용한다.
- 사후 분포 추정의 결과는 히스토그램으로 확인할 수 있다.

### 8. **사후 분포 추론**

- 깁스 샘플링을 통해 추출된 샘플들로부터 사후 평균과 분산을 추정할 수 있다:
  $
  E[\beta_j | Y] \approx \frac{1}{n_1} \sum_{j=1}^{n_1} \beta_j^{(i)}
  $
  $
  Var(\beta_j | Y) \approx \frac{1}{n_1} \sum_{j=1}^{n_1} (\beta_j^{(i)} - E[\beta_j])^2
  $



#### 사후분포
관측된 데이터를 통해 업데이트된 $\beta$와 $\sigma^2$의 사후 분포는 다음과 같이 구할 수 있습니다. 데이터 $Y$가 주어지면 $\beta$와 $\sigma^2$는 다음과 같은 사후 분포를 따릅니다.

- $\beta$의 사후 분포:
  $
  \pi(\beta|Y) \sim Normal(B_1 A, B_1)
  $
    - $B_1 = (X'X + B_0^{-1})^{-1}$, 
    - $A = X'Y + B_0^{-1}\beta_0$

- $\sigma^2$의 사후 분포:
  $
  \sigma^2 \sim InverseGamma\left(\frac{\alpha_0 + T}{2}, \frac{\delta_0 + (Y - X\beta)'(Y - X\beta)}{2}\right)
  $

이를 통해 우리는 회귀 계수 $\beta$와 잔차 분산 $\sigma^2$에 대해 데이터 기반으로 추론할 수 있으며, 사전 정보가 데이터에 반영된 결과를 얻게 됩니다.


```python
import pymc as pm
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import pandas as pd
import platform


# 한글 폰트 설정
if platform.system() == 'Darwin':  
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':  
    rc('font', family='Malgun Gothic')
else:  
    rc('font', family='NanumGothic')
plt.rcParams['axes.unicode_minus'] = False
```

```python
quarterly_data = pd.read_csv('data/quarterly_combined.csv', index_col=0, parse_dates=True).astype(float)
quarterly_data
```

```python
# 환율 변화율 (S_t+1 - S_t) / S_t
exchange_rate_change = quarterly_data.usd_krw_quarterly.pct_change().iloc[1:]
# 국내 이자율 (i_t)
domestic_interest_rate = quarterly_data.korea_cd_quarterly.iloc[1:]
# 미국 이자율 (i_t^*)
foreign_interest_rate = quarterly_data.treasury_3month_quarterly.iloc[1:]
```

```python
# 내외 금리차 계산
interest_rate_diff = domestic_interest_rate - foreign_interest_rate

# 그림을 그릴 날짜 범위 선택
dates = exchange_rate_change.index

# 그래프 그리기
fig, ax1 = plt.subplots(figsize=(10, 6))

# 좌측 y축: 환율 변화율
ax1.plot(dates, exchange_rate_change * 100, 'b--', label='depreciation rate (LHS)')
ax1.set_ylabel('Depreciation rate (%)', color='b')  # 좌측 y축 레이블 설정
ax1.tick_params(axis='y', labelcolor='b')  # 좌측 y축 눈금 색상 설정3 

# 우측 y축: 내외 금리차
ax2 = ax1.twinx()  # 두 번째 y축 생성
ax2.plot(dates, interest_rate_diff, 'k-', label='i_t - i_t^* (RHS)')
ax2.set_ylabel('Interest Rate Difference (%)', color='k')  # 우측 y축 레이블 설정
ax2.tick_params(axis='y', labelcolor='k')  # 우측 y축 눈금 색상 설정

# 그래프 제목 및 범례 설정
plt.title('환율변화율과 내외 금리차')  # 그래프 제목 설정
ax1.legend(loc='upper left')  # 좌측 y축 범례 위치 설정
ax2.legend(loc='upper right')  # 우측 y축 범례 위치 설정

# 그래프 출력
plt.show()  # 그래프 표시
```

```python
daily_data = pd.read_csv('data/daily_combined.csv', index_col=0, parse_dates=True).astype(float)
daily_data
```

```python
exchange_rate_change_daily = daily_data.usd_krw_daily.pct_change().iloc[1:]
domestic_interest_rate_daily = daily_data.korea_cd_daily.iloc[1:]
foreign_interest_rate_daily = daily_data.treasury_3month_daily.iloc[1:]
```

```python
interest_rate_diff_daily = domestic_interest_rate_daily - foreign_interest_rate_daily
dates_daily = exchange_rate_change_daily.index
# 그래프 그리기
fig, ax1 = plt.subplots(figsize=(10, 6))

# 좌측 y축: 환율 변화율

ax1.plot(dates_daily, exchange_rate_change_daily * 100, 'b--', label='depreciation rate (LHS)')
ax1.set_ylabel('Depreciation rate (%)', color='b')
ax1.tick_params(axis='y', labelcolor='b')

# 우측 y축: 내외 금리차
ax2 = ax1.twinx()
ax2.plot(dates_daily, interest_rate_diff_daily, 'k-', label='i_t - i_t^* (RHS)')
ax2.set_ylabel('Interest Rate Difference (%)', color='k')
ax2.tick_params(axis='y', labelcolor='k')

# 그래프 제목 및 범례 설정
plt.title('환율변화율과 내외 금리차')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# 그래프 출력
plt.show()

```

```python
T = len(exchange_rate_change)
# 2. 선형 회귀 모형 설정
with pm.Model() as model:
    # 사전 분포 설정
    sigma2 = pm.InverseGamma('sigma2', alpha=25, beta=12500)
    beta = pm.MvNormal('beta', mu=np.zeros(2), cov=25 * np.eye(2), shape=2)

    # 설명변수 X 구성
    X = np.column_stack([
        np.ones(T),  # 상수항 a1
        domestic_interest_rate - foreign_interest_rate  # i_t - i_t^*
    ])

    # 종속변수 Y
    Y = exchange_rate_change

    # 우도 함수 설정 (선형 회귀)
    Y_obs = pm.Normal('Y_obs', mu=pm.math.dot(X, beta), sigma=np.sqrt(sigma2), observed=Y)

    # 5000번 샘플링
    trace = pm.sample(5000, cores=4, chains=4)

# 추정 결과 시각화
pm.plot_trace(trace)
plt.show()

# 추정된 값 요약
summary = pm.summary(trace)
print(summary)
```
